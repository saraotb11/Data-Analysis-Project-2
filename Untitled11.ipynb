```python
def clean_text(text):
    # تحويل جميع النصوص إلى حروف صغيرة
    text = text.lower()

    # إزالة الأرقام من النص
    text = re.sub(r'\d+', '', text)

    # إزالة علامات الترقيم من النص
    text = text.translate(str.maketrans('', '', string.punctuation))

    # إزالة أي مسافات بيضاء في بداية ونهاية النص
    text = text.strip()

    # إرجاع النص المنظف
    return text
```

```python
data['cleaned_text'] = data['Text'].apply(clean_text)
```

```python
# تصفية الدرجات المحايدة (Score 3)
df = data[data['Score'] != 3]

# تحويل الدرجات 1 و 2 إلى 0 (سلبية)، والدرجات 4 و 5 إلى 1 (إيجابية)
df['sentiment'] = df['Score'].apply(lambda x: 1 if x > 3 else 0)

# حذف الأعمدة غير الضرورية من مجموعة البيانات
df = df.drop(columns=['Id', 'ProductId', 'UserId', 'ProfileName', 'HelpfulnessNumerator', 'HelpfulnessDenominator', 'Score', 'Time'])
```

```python
#  انشاء كائن عداد لتخزين عدد مرات ظهور الكلمات(Counter)
count = Counter()


for text in df["text"].values:
    for word in text.split():
        count[word] += 1  # زيادة عدد ظهور الكلمة بمقدار 1

# الحصول على أكثر 10 كلمات شيوعًا
count.most_common(10)

# إنشاء DataFrame
temp = pd.DataFrame(count.most_common(10))

# تسمية الأعمدة في DataFrame
temp.columns = ['word', 'count']

# عرض  الناتج
temp
```

```python
# إنشاء مخطط شريطي باستخدام Plotly لإظهار الكلمات الشائعة في النصوص
fig = px.bar(temp, x="count", y="word", title='Common Words in Text', orientation='h',
             width=700, height=700)

# تعريف مجموعة من الألوان لاستخدامها في المخطط
colors = ['#636EFA', '#EF553B', '#00CC96', '#AB63FA', '#FFA15A',
          '#19D3F3', '#FF6692', '#B6E880', '#FF97FF', '#FECB52']

# تحديث لون الأشرطة في المخطط بالألوان المحددة
fig.update_traces(marker_color=colors)

# عرض المخطط الناتج
fig.show()
```

```python
# استخراج القيم من عمود
x = df['text'].values
y = df['sentiment'].values

# إنشاء متجهات من النصوصTF-IDF
word_vector = TfidfVectorizer()
x = word_vector.fit_transform(x)  # تحويل النصوص إلى متجهات TF-IDF
```

```python
# تقسيم البيانات إلى مجموعة تدريب (80%) ومجموعة اختبار (20%)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
# إنشاء نموذج Complement Naive Bayes
CNB = ComplementNB()

# تدريب النموذج باستخدام مجموعة التدريب
CNB.fit(x_train, y_train)
```

```python

from sklearn import metrics

predicted = CNB.predict(x_test)

# حساب دقة النموذج
accuracy_score = metrics.accuracy_score(predicted, y_test)

# طباعة دقة النموذج
print('ComplementNB model accuracy is', str('{:04.2f}'.format(accuracy_score * 100)) + '%')

# طباعة مصفوفة الارتباك (Confusion Matrix)
print('Confusion Matrix:')
print(pd.DataFrame(confusion_matrix(y_test, predicted)))

# طباعة تقرير التصنيف (Classification Report)
print('Classification Report:')
print(classification_report(y_test, predicted))
```

```python
# إنشاء نموذج LogisticRegression
model = LogisticRegression()

# تدريب النموذج
model.fit(x_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, classification_report

# إجراء التوقعات باستخدام مجموعة الاختبار
y_pred = model.predict(x_test)

# تقييم النموذج
print(f'Accuracy: {accuracy_score(y_test, y_pred)}')  # طباعة دقة النموذج

# طباعة تقرير التصنيف (Classification Report) لتحليل الأداء
print(classification_report(y_test, y_pred))
```

```python
from sklearn.naive_bayes import MultinomialNB

# إنشاء نموذج Naive Bayes متعدد الحدود
MNB = MultinomialNB()

# تدريب النموذج باستخدام مجموعة التدريب
MNB.fit(x_train, y_train)
```

```python
# إجراء التوقعات باستخدام مجموعة الاختبار
predicted = MNB.predict(x_test)

# حساب دقة النموذج
accuracy_score = metrics.accuracy_score(predicted, y_test)

# طباعة دقة النموذج
print('MultinominalNB model accuracy is', str('{:04.2f}'.format(accuracy_score * 100)) + '%')
print('------------------------------------------------')

# طباعة مصفوفة الارتباك (Confusion Matrix)
print('Confusion Matrix:')
print(pd.DataFrame(confusion_matrix(y_test, predicted)))
print('------------------------------------------------')

# طباعة تقرير التصنيف (Classification Report)
print('Classification Report:')
print(classification_report(y_test, predicted))
```
